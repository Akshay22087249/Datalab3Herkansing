```python
# Een lijst maken van de csv om ze in één keer in een for loop te behandelen:
name_list = ['Delft', 'Huisvesting_Delft', 'Poseidon', 'Rugzak', 'Slinger', 
             'Sportcampus', 'Strip', 'Zoetermeer', 'Zwembad']

csv_list = ["delft_timetable.csv", "huisvesting_delft_timetable.csv", 
            "poseidon_timetable.csv", "rugzak_timetable.csv", 
            "slinger_timetable.csv", "sportcampus_zuiderpark_timetable.csv", 
            "strip_timetable.csv", "zoetermeer_timetable.csv"]

cleaned_dfs = {}

for name, csv in zip(name_list, csv_list):
    df = pd.read_csv(csv)
    df = pd.DataFrame(df)
    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
    df = df.dropna(how='all')
    df = df.drop(['module_code', 'start_week_(iso)', 'duration', 'type', 'student_set(s)', 'department', 'size', 'attendance_is_mandatory_for_this_activity', 'check', 'id', 'start_week'], axis=1)
    df = df.sort_values('start_date')
    cleaned_dfs[f"{name}_clean"] = df

cleaned_dfs['Zoetermeer_clean']
```
