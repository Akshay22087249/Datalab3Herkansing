```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Stap 1: Start browser
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://mytimetable.hhs.nl/schedule")

# Stap 2: Wacht tot de loginpagina geladen is
time.sleep(3)

# Stap 3: Vul je login in (afhankelijk van het type login)
username = driver.find_element(By.NAME, "username")  # pas deze 'name' aan op basis van de HTML
password = driver.find_element(By.NAME, "password")  # idem
username.send_keys("jouw_gebruikersnaam")
password.send_keys("jouw_wachtwoord")
password.send_keys(Keys.RETURN)

# Stap 4: Wacht tot de timetable is geladen
time.sleep(5)

# Stap 5: Haal de HTML op
html = driver.page_source

# Optioneel: gebruik BeautifulSoup om specifieke data te extraheren
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Bijvoorbeeld alle vaknamen of lokaalnummers zoeken
for event in soup.find_all('div', class_='event'):
    print(event.text)

# Stap 6: Sluit browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Chrome()
driver.get("https://mytimetable.hhs.nl/schedule")

# Handmatig inloggen + authenticator
input("Log in en druk op Enter als de timetable zichtbaar is...")

try:
    # Wacht tot het rooster geladen is (pas selector aan!)
    timetable = WebDriverWait(driver, 15).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "div.timetable"))
    )
    print(timetable.text)  # Of sla data op in een bestand
except Exception as e:
    print("Fout:", e)
finally:
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Chrome()
driver.get("https://mytimetable.hhs.nl/schedule")

# Stap 1: Handmatig inloggen + Authenticator
input("Log in handmatig en druk op Enter als het rooster zichtbaar is...")

try:
    # Stap 2: Wacht tot de pagina laadt
    WebDriverWait(driver, 20).until(
        EC.presence_of_element_located((By.CLASS_NAME, "week-navigation"))  # Pas aan!
    )

    # Stap 3: Rooster uitlezen
    timetable = driver.find_element(By.CLASS_NAME, "timetable-grid")  # Pas aan!
    print("Rooster gevonden! Eerste tekst:", timetable.text[:200])  # Print eerste 200 chars

except Exception as e:
    print("Fout bij scrapen:", e)

finally:
    driver.quit()
```

```python
from bs4 import BeautifulSoup

with open("timetable.html", "r", encoding="utf-8") as f:
    soup = BeautifulSoup(f, "html.parser")

roosters = []

for div in soup.find_all("div", style=True):
    style = div['style']
    content = div.get_text(strip=True)

    if "top:" in style and "left:" in style:
        roosters.append({
            "style": style,
            "content": content
        })

for item in roosters:
    print(item)
```

```python
from bs4 import BeautifulSoup

with open("rooster_pagina.html", "r", encoding="utf-8") as f:
    soup = BeautifulSoup(f, "lxml")

events = soup.find_all("div", class_="event")

rooster_data = []

for event in events:
    style = event.get("style", "")
    content = event.get_text(separator=" | ", strip=True)

    # Optioneel: extract style positioning if needed
    event_info = {
        "style": style,
        "inhoud": content
    }

    rooster_data.append(event_info)

# Print of bewaar
for item in rooster_data:
    print(item["inhoud"])
```
